```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
# read the dataset

df = pd.read_csv('/content/emp_promotion.csv')
df.head()
```

```python
# Checking the dtype

df.info()
```

```python
# Finding the sum of null values

df.isnull().sum()
```

```python
# Checking the shape of the data

df.shape
```

```python
# Dropping the unwanted column

df.drop('employee_id',axis=1,inplace=True)

# df = df.drop('employee_id',axis=1)
```

```python
df.info()
```

```python
df['education'].mode()[0]
```

```python
df['previous_year_rating'].mean(), df['previous_year_rating'].median()
```

```python
df.describe()
```

```python
# Replacing null values

df['education'] = df['education'].fillna(df['education'].mode()[0])
df['previous_year_rating'] = df['previous_year_rating'].fillna(df['previous_year_rating'].mean())
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df['department'] = le.fit_transform(df['department'])
df['region'] = le.fit_transform(df['region'])
df['education'] = le.fit_transform(df['education'])
df['gender'] = le.fit_transform(df['gender'])
df['recruitment_channel'] = le.fit_transform(df['recruitment_channel'])
```

```python
df.head()
```

```python
# Split the data

from sklearn.model_selection import train_test_split
```

```python
# Indep and dep variable split

x = df.drop('is_promoted',axis=1)
y = df['is_promoted']
```

```python
# Training and testing data

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=12)
```

```python
X_train.shape, X_test.shape
```

```python
# Initialize the model

knn = KNeighborsClassifier()
```

```python
# Train the model

knn.fit(X_train,y_train)
```

```python
ypred = knn.predict(X_test)
```

```python
ypred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,ypred))
```

```python
confusion_matrix(y_test,ypred)
```
